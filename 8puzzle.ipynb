```python
class Node:
    def __init__(self,state,parent,level):
        self.state=state
        self.parent=parent
        self.level=level
```

BFS Algorithm

```python
from collections import deque
import time 
def bfs(initial_state):
    goal_state = [0, 1, 2, 3, 4, 5, 6, 7, 8]

    frontier = deque()  # Create a deque to store the Nodes(states) to be explored
    frontier.append(Node(initial_state,None,0)) # create first Node(state,parent,level)

    frontier_states = set()  # Create a set to keep track of states in the frontier (set is O(1) AvgCase Search)
    frontier_states.add(tuple(initial_state)) #set must take mutable items (tuple or string) so we change list to tuple

    explored = set() # Create a set to keep track of explored states

    start_time = time.time() # Record the start time for measuring execution time
    search_depth = 0 # Initialize the search depth

    while frontier: # Continue searching as long as there are states in the frontier
        current_node = frontier.popleft()  # Get the Node from the front of the frontier

        search_depth = max(search_depth, current_node.level)  #lazm yeb2a hena mynzlsh tht 3shan lw l2a goal msh hy3del l depth 

        if current_node.state == goal_state: # Check if the current state is the goal state
            end_time = time.time() # Record the end time
            execution_time = end_time - start_time # Calculate the running time

            node=current_node
            path=[]
            while node.parent is not None:
                path.append(node.state)
                node=node.parent

            return path[::-1], len(path), len(explored), search_depth, execution_time  

        explored.add(tuple(current_node.state))  

        empty_tile_index = current_node.state.index(0) # Find the index of the empty tile (0)
        x, y = empty_tile_index // 3, empty_tile_index % 3 # Calculate its coordinates in the 3x3 grid

        # Define possible moves: Up, Down, Left, Right
        moves = [(x - 1, y), (x + 1, y), (x, y - 1), (x, y + 1)]

        for move_x, move_y in moves:   
            if 0 <= move_x < 3 and 0 <= move_y < 3: # Check if the move is within the grid
                new_state = current_node.state[:]  # Create a copy of the current state
                new_empty_tile_index = move_x * 3 + move_y  # Calculate the index of the new empty tile
                new_state[empty_tile_index], new_state[new_empty_tile_index] = new_state[new_empty_tile_index], new_state[empty_tile_index]  

                new_state_tuple = tuple(new_state) # Convert the new state to a tuple 

                # Check if the new state has not been explored and is not in the frontier
                if new_state_tuple not in explored and new_state_tuple not in frontier_states:
                    frontier.append(Node(new_state,current_node,current_node.level+1))  
                    frontier_states.add(new_state_tuple) # Add the new state to the set of frontier states 
```

```python
from queue import PriorityQueue
from math import sqrt

def manhattan_distance(state):
    distance = 0
    for i in range(9):
        tile = state[i]
        if tile != 0:
            goal_row, goal_col = divmod(tile, 3)#btrg3 div w mod
            current_row, current_col = divmod(i, 3)
            distance += abs(current_row - goal_row) + abs(current_col - goal_col)
    return distance

def euclidean_distance(state):
    distance = 0
    for i in range(9):
        tile = state[i]
        if tile != 0:
            goal_row, goal_col = divmod(tile, 3)
            current_row, current_col = divmod(i, 3)
            dx = current_row - goal_row
            dy = current_col - goal_col
            distance += sqrt(dx * dx + dy * dy)
    return distance
```

#### Determine whether the initial puzzle state is solvable or not!

```python
# Count inversions to determine whether the initial puzzle state is solvable or not
def count_inversions(puzzle):
    flat_puzzle = [tile for tile in puzzle if tile != 0]
    inversions = 0

    for i in range(len(flat_puzzle)):
        for j in range(i + 1, len(flat_puzzle)):
            if flat_puzzle[i] > flat_puzzle[j]:
                inversions += 1
    return inversions

# Function to check if a puzzle is solvable
def is_solvable(puzzle):
    inversions = count_inversions(puzzle)
    return inversions % 2 == 0
```
